```python
from typing import List

class Solution:
    def hasDuplicate(self, nums: List[int]) -> bool:
        check = set()
        for num in nums:
            if num in check:
                return True
            check.add(num)
        return False
```

```python
sol = Solution()
print(sol.hasDuplicate([1,2,3,1]))   
print(sol.hasDuplicate([1,2,3,4]))   
```
