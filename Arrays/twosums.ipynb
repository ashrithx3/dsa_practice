```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:

        dic = {}
        for i,j in enumerate(nums):
            reverse = target - j
            if reverse in dic:
                if i > dic[reverse]:
                    return [dic[reverse],i]
                else:
                    return[i,dic[reverse]]
            else:
                dic[j] = i
```
